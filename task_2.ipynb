```python
def area_of_circle(radius):
    area = 3.14159 * radius * radius
    return area

# Example
print(area_of_circle(5))
```

```python
def largest(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

# Example
print(largest(10, 25, 15))
```

```python
def string_length(text):
    count = 0
    for char in text:
        count += 1
    return count

# Example
print(string_length("Python"))
```

```python
def is_prime(num):
    if num <= 1:
        return False

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False

    return True

# Example
print(is_prime(17))
```

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

# Example
print(fibonacci(6))
```

```python
def reverse_string(text):
    if text == "":
        return ""
    return reverse_string(text[1:]) + text[0]

# Example
print(reverse_string("Python"))
```

```python
n = int(input("Enter a number: "))

total = 0

for i in range(1, n + 1):
    total += i

print("Sum =", total)
```

```python
numbers = [10, 15, 22, 33, 40, 51, 60]

count = 0

for num in numbers:
    if num % 2 == 0:
        count += 1

print("Even numbers =", count)
```

```python
for num in range(2, 101):
    prime = True

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            prime = False
            break

    if prime:
        print(num, end=" ")
```
